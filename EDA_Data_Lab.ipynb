### **What** : Apa yang membuat karyawan meninggalkan perusahaan?

````python
# Membuat list 'what'
what = ['OverTime', 'JobInvolvement', 'BusinessTravel']

# Membuat subplots
fig, ax = plt.subplots(1,3, figsize = (18,4))

# Menghitung proporsi attrition untuk list 'what'
for i in range(len(what)):
  df_what = df.groupby([what[i]])['Attrition'].value_counts(normalize = True).unstack()
  df_what = df_what[['Yes', 'No']]
  df_what = df_what.sort_values('Yes', ascending = False)

  # Membuat barchart
  ax1 = df_what.plot.bar(stacked=True, color=['#dd4124', '#00bfc4'], edgecolor = 'black', linewidth = 0.5, ax = ax[i])
  if i == 0:
    ax1.set_ylabel('Percentage %')
  ax1.set_xticklabels(df_what.index, rotation = 0)

# Menampilkan grafik
plt.show()
````

Setelah dilakukan proses analisa, didapatkan beberapa kesimpulan berupa :
1. Karyawan yang bekerja melebihi jam kerja reguler (*over time*) memiliki *Attrition Rate* yang lebih tinggi dibandingkan dengan karyawan yang tidak bekerja lembur.
2. Karyawan yang memiliki keterlibatan kerja yang lebih rendah memiliki *Attrition Rate* yang lebih tinggi.
3. Karyawan yang lebih sering melakukan *business travel* memiliki *Attrition Rate* yang lebih tinggi.


### **Who** : Siapa saja yang karyawan yang meninggalkan perusahaan?

````python
# Membuat list 'who'
who = ['Age', 'Gender', 'JobLevel', 'MaritalStatus', 'Education', 'EducationField', 'JobRole']

# Membuat subplots
fig, ax = plt.subplots(4,2, figsize = (15,15))

# Menghitung proporsi attrition untuk list 'who'
for i in range(len(who)):
  df_who = df.groupby([who[i]])['Attrition'].value_counts(normalize = True).unstack()
  df_who = df_who[['Yes', 'No']]
  df_who = df_who.sort_values('Yes', ascending = False)

  # Membuat histogram dan kdeplot
  if i == 0:
    sns.histplot(df, x = df['Age'], hue = 'Attrition', kde = True, palette = ['#dd4124', '#00bfc4'], edgecolor = 'black', linewidth = 0.5, alpha = 1, ax=ax[0,0])
    ax1 = sns.kdeplot(x = df[df['Attrition'] == 'Yes']['Age'], fill = True, label = 'Attrition', color = '#dd4124', edgecolor = 'black', linewidth = 1, alpha = 0.9, ax = ax[0,1])
    ax2 = sns.kdeplot(x = df[df['Attrition'] == 'No']['Age'], fill = True, label = 'No Attrition', color = '#00bfc4', edgecolor = 'black', linewidth = 1, alpha = 0.9, ax = ax[0,1])

    # Menampilkan legenda
    ax1.legend()

  # Membuat barchart
  elif i in [1, 2]:
    bar_who = df_who.plot.bar(stacked=True, color=['#dd4124', '#00bfc4'], edgecolor = 'black', linewidth = 0.5, ax=ax[1,i-1])
    bar_who.set_ylabel('Percentage %')
    bar_who.set_xticklabels(df_who.index, rotation = 0)
  elif i in [3, 4]:
    bar_who = df_who.plot.bar(stacked=True, color=['#dd4124', '#00bfc4'], edgecolor = 'black', linewidth = 0.5, ax=ax[2,i-3])
    bar_who.set_ylabel('Percentage %')
    bar_who.set_xticklabels(df_who.index, rotation = 0)
  else:
    df_who = df_who.sort_values('Yes')
    bar_who = df_who.plot.barh(stacked=True, color=['#dd4124', '#00bfc4'], edgecolor = 'black', linewidth = 0.5, ax=ax[3,i-5])
    bar_who.set_xlabel('Percentage %')
  plt.xticks(rotation = 0)

# mengatur jarak antar grafik
plt.tight_layout()

# Menampilkan grafik
plt.show()
````

Berdasarkan grafik di atas dapat disimpulkan bahwa :
1. Karyawan dengan rentang umur di bawah 40 tahun cenderung memiliki *Attrition Rate* yang lebih tinggi. Kemudian, karyawan dengan rentang umur sekitar 25 - 35 tahun memiliki *Attrition rate* yang paling tinggi.
2. Karyawan pria cenderung memiliki *Attrition Rate* yang relatif sama dengan karyawan wanita.
3. Karyawan yang memiliki Job Level yang lebih rendah cenderung memiliki attration rate yang lebih tinggi. Kemudian, karyawan yang memiliki Job Level 1 merupakan tipe karyawan dengan *Attrition Rate* paling tinggi.
4. Karyawan dengan marital status belum menikah (*single*) cenderung memiliki *Attrition Rate* yang lebih tinggi dibandingkan karyawan  yang sudah menikah dan yang berstatus sudah bercerai. Selain itu, karyawan dengan *marital status* *divorced* memiliki *Attrition Rate* paling rendah atau merupakan tipe karyawan yang paling banyak bertahan di perusahaan.
5. Karyawan yang memiliki educational level yang lebih rendah cenderung memiliki *Attrition rate* yang lebih tinggi. Kemudian, karyawan dengan yang memiliki *educational level* 1 merupakan tipe karyawan yang memiliki *Attrition Rate* paling tinggi.
6. Karyawan yang memiliki latar belakang pendidikan dibidang Human Resources, Technical Degree, dan Marketing merupakan 3 golongan karyawan yang memiliki attration rate paling tinggi.
7. Karyawan yang menjabat sebagai *Sales Representative*, *Laboratory Technician*, dan *Human Resources* merupakan 3 golongan karyawan yang memiliki *Attrition rate* paling tinggi.


### **Why** : Alasan apa yang membuat karyawan yang meninggalkan perusahaan?

````python
# Membuat list 'why'
why =  ['MonthlyIncome', 'EnvironmentSatisfaction', 'JobSatisfaction']

# Membuat subplots
fig, ax = plt.subplots(2,2, figsize = (15,8))

# Membuat kdeplot
ax1 = sns.kdeplot(x = df[df['Attrition'] == 'Yes']['MonthlyIncome'], fill = True, label = 'Attrition', color = '#dd4124', edgecolor = 'black', linewidth = 1, alpha = 0.9, ax = ax[0,0])
ax2 = sns.kdeplot(x = df[df['Attrition'] == 'No']['MonthlyIncome'], fill = True, label = 'No Attrition', color = '#00bfc4', edgecolor = 'black', linewidth = 1, alpha = 0.9, ax = ax[0,0])

# Membuat barchart jobrole - monthly income
jobrole_income = df.groupby("JobRole")["MonthlyIncome"].mean().to_frame()
jobrole_income.columns = ['Mean MonthlyIncome']
jobrole_income = jobrole_income.sort_values('Mean MonthlyIncome')
jobrole_income.plot.barh(color=['#00bfc4'], edgecolor = 'black', linewidth = 0.5, ax=ax[0,1])

# Menampilkan legenda
ax1.legend()

# Menghitung proporsi attrition untuk list 'why'
for i in range(1, len(why)):
  df_why = df.groupby([why[i]])['Attrition'].value_counts(normalize = True).unstack()
  df_why = df_why[['Yes', 'No']]
  df_why = df_why.sort_values('Yes', ascending = False)

  # Membuat barchart
  bar_why = df_why.plot.bar(stacked = True, color = ['#dd4124', '#00bfc4'], edgecolor = 'black', linewidth = 0.5, ax = ax[1,i-1])
  bar_why.set_ylabel('Percentage %')
  bar_why.set_xticklabels(df_why.index, rotation = 0)

# mengatur jarak antar grafik
plt.tight_layout()

# Menampilkan grafik
plt.show()
````

Berdasarkan grafik di atas dapat disimpulkan bahwa :

### **Where** : Dari mana sajakah karyawan yang meninggalkan perusahaan?


````python
# Membuat list 'where'
where =  ['Department']

# Menghitung proporsi attrition untuk list 'where'
for i in range(len(where)):
  df_where = df.groupby([where[i]])['Attrition'].value_counts(normalize = True).unstack()
  df_where = df_where[['Yes', 'No']]
  df_where = df_where.sort_values('Yes', ascending = False)

  # Membuat barchart
  df_where.plot.bar(stacked=True, color=['#dd4124', '#00bfc4'], edgecolor = 'black', linewidth = 0.5)
  plt.ylabel('Percentage %')
  plt.xticks(rotation = 0)

# Menampilkan grafik
plt.show()
````

Berdasarkan grafik di atas dapat disimpulkan bahwa karyawan yang bekerja pada Departemen Sales memiliki *Attrition Rate* yang paling tinggi dibandingkan dengan departemen lainnya.
